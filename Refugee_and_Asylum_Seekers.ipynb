```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset
df = pd.read_csv("demo.csv")

# Removeing the first row
df_clean = df.iloc[1:].copy()

# Converting 'Year' and 'Total' columns to numeric
df_clean['Year'] = pd.to_numeric(df_clean['Year'], errors='coerce')
df_clean['Total'] = pd.to_numeric(df_clean['Total'], errors='coerce')

# Droping rows with missing values in 'Year' or 'Total'
df_clean = df_clean.dropna(subset=['Year', 'Total'])

# Creating the scatterplot
plt.figure(figsize=(10, 6))
plt.scatter(df_clean['Year'], df_clean['Total'], alpha=0.6)
plt.title('Total Affected Individuals Over Time')
plt.xlabel('Year')
plt.ylabel('Total Affected')
plt.grid(True)
plt.tight_layout()
plt.show()
# Creating The Barplot
plt.bar(df_clean['Year'], df_clean['Total'])
plt.title('Total Affected Individuals Over Time')
plt.xlabel('Year')
plt.ylabel('Total Affected')
plt.grid(True)
plt.tight_layout()
plt.show()
```
